```python
a = [1, 3, 2, 8]
x = np.array(a)
print(f'{x=}')
print(f'{type(x)=}')   # numpy.ndarray je obecné n-rozměrné pole
print(f'{x.size=}')  # celková velikost
print(f'{x.shape=}')  # rozměry v jednotlivých dimenzích
print(f'{x.dtype=}')  # datový typ položek
```

```python
a = -3
while True:
    if a == 3:
        print(f'continue at {a=}')
        a += 2 
        continue
    if a > 4:
        print(f'break {a=}')
        break
    a += 1
    print(a)
```

```python
# structural pattern matching - tak trochu navíc


# x = 4
# x = 9
# x = [9]
# x = [1, 7]
x = {'a': 4, 'b': 8}

match x:
    case 4:
        print('ctyrka')
    case (i,):
        print(f'jedna polozka {i=}')
    case (p, q):
        print(f'dve polozky: {p} a {q}')
    case _:
        print('neco jineho')
```

```python
# graf funkce s matplotlib

x = np.linspace(-2*np.pi, 4*np.pi, 1000)

y1 = 1.5 * np.sin(x) + 0.1  # vektorizovane funkce z numpy
y2 = np.cos(x**2)

plt.plot(x, y1)
plt.plot(x, y2)
plt.grid()
plt.show()
```
